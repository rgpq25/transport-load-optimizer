```python
"""
This script reads the CSV grasp_debug_log.csv file generated by the algorithm.
"""

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "../output/saga_debug_log.csv"
df = pd.read_csv(file_path)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Temperatura'], df['Fitness promedio'], marker='o')
plt.gca().invert_xaxis()  # Descending temperature on X axis
plt.title('Promedio de Fitness de la Población vs Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Fitness Promedio de la Población')
plt.grid(True)
plt.tight_layout()
plt.show()
```
