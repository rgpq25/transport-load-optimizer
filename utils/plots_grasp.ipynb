```python
"""
This script reads the CSV grasp_debug_log.csv file generated by the algorithm.
"""

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "../output/grasp_debug_log.csv"
df = pd.read_csv(file_path)

step = 10
colName = 'Mejor fitness global' # 'Mejor fitness global' | 'Fitness patrón generado'

# Plotting
df = df.iloc[::step, :]
plt.figure(figsize=(10, 6))
plt.plot(df['Iteración'], df[colName], marker='o')
# plt.gca().invert_xaxis()  # Descending temperature on X axis
# plt.title('Fitness patrón generado vs Iteración')
plt.xlabel('Iteración')
plt.ylabel(colName)
plt.grid(True)
plt.tight_layout()
plt.show()
```
